    Q1. Ans CALCULATING FACTORIAL OF GIVEN NO

```python
def factorial_of_no(x):
    if x == 1:
        return x
    elif x < 0:
        return ("Not available")
    elif x == 0:
        return 1
    else:
        return x*factorial_of_no(x-1)
print (factorial_of_no)
```

```python
factorial_of_no(5)
```
